```python
# Read in all conference pitch logs and concatenate them to create one big csv file. Save locally, file too big to push to GitHub
```

```python
import pandas as pd
import glob

files = glob.glob("pitch_logs/**/*.csv", recursive=True)
df = pd.concat([pd.read_csv(f) for f in files], ignore_index=True)
```

```python
del df["Unnamed: 0"]
```

```python
len(df)
```

```python
df.to_csv("full_data.csv")
```
